```python
import json
import glob
```

```python
running_id = 0
missing_json_list = []
completed_json_list = []
gt_json_list = []


for name in glob.glob('*.json'):
    print(running_id)
    print(name)
    
    flag = 0
    with open(name) as f:
        jsonparsedfull = json.load(f)
        path = jsonparsedfull['path']
        height = jsonparsedfull['height']
        width = jsonparsedfull['width']
        
    #predicate check
    preds = [obj['predicate'] for obj in jsonparsedfull['missing_scene']['relationships']] + [obj['predicate'] for obj in jsonparsedfull['complete_scene']['relationships']] + [obj['predicate'] for obj in jsonparsedfull['gt_scene']['relationships']]
    for pred in preds:
        if(pred in ['ride', '4', 4]):
            flag = 1
            break
            
    if(flag==1):
        continue
            
        
    # Missing scene
    jsonparsed = jsonparsedfull['missing_scene']
    jsonparsed['height'] = height
    jsonparsed['width'] = width
    jsonparsed['regions'] = []
    jsonparsed['path'] = path
    jsonparsed['id'] = running_id
    
    for idx, obj in enumerate(jsonparsed['objects']):
        jsonparsed['objects'][idx]['box'] = [0,0,0,0]
    
    missing_json_list.append(jsonparsed)
    
    # Completed scene
    jsonparsed_comp = jsonparsedfull['complete_scene']
    jsonparsed_comp['height'] = height
    jsonparsed_comp['width'] = width
    jsonparsed_comp['regions'] = []
    jsonparsed_comp['path'] = path
    jsonparsed_comp['id'] = running_id
    
    for idx, obj in enumerate(jsonparsed_comp['objects']):
        jsonparsed_comp['objects'][idx]['box'] = [0,0,0,0]
    
    completed_json_list.append(jsonparsed_comp)
    
    # GT scene
    jsonparsed_gt = jsonparsedfull['gt_scene']
    jsonparsed_gt['height'] = height
    jsonparsed_gt['width'] = width
    jsonparsed_gt['regions'] = []
    jsonparsed_gt['path'] = path
    jsonparsed_gt['id'] = running_id
    
#     for idx, obj in enumerate(jsonparsed_gt['objects']):
#         jsonparsed_gt['objects'][idx]['box'] = [0,0,0,0]
    
    gt_json_list.append(jsonparsed_gt)
    
    
    running_id = running_id + 1
```

```python
with open("finaldata_missing.json", "w") as f:
    json.dump(missing_json_list, f)
```

```python
with open("finaldata_gt.json", "w") as f:
    json.dump(gt_json_list, f)
```
